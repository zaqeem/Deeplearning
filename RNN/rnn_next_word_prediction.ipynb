```python
from datasets import load_dataset

# Load the dataset
dataset = load_dataset("wikitext", "wikitext-2-raw-v1")

# Extract training text
train_texts = dataset["train"]["text"]

# Combine all text into a single string
text_data = " ".join(train_texts)

print("Total length of text:", len(text_data))
print("Sample text:", text_data[:500])  # Print first 500 characters
```

```python
from transformers import AutoTokenizer

# Load a pre-trained tokenizer (GPT-2 tokenizer)
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# Tokenize the text (convert words into numbers)
tokenized_text = tokenizer(text_data, add_special_tokens=False)["input_ids"]

print("Sample tokenized output:", tokenized_text[:20])  # Print first 20 token IDs
print("Vocabulary size:", tokenizer.vocab_size)
```

```python
import torch

# Convert token list to tensor
token_tensor = torch.tensor(tokenized_text, dtype=torch.long)

print("Token tensor shape:", token_tensor.shape)  # Shape: (total tokens,)
```

```python
import torch.nn as nn

# Define Simple RNN Model
class SimpleRNNModel(nn.Module):
    def __init__(self, vocab_size, embed_size, hidden_size):
        super(SimpleRNNModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_size)
        self.rnn = nn.RNN(embed_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, vocab_size)

    def forward(self, x):
        x = self.embedding(x)  # Convert token IDs to word embeddings
        x, _ = self.rnn(x)  # Process sequence using Simple RNN
        x = self.fc(x[:, -1, :])  # Take the last output for prediction
        return x

# Model Parameters
VOCAB_SIZE = tokenizer.vocab_size  # Size of vocabulary
EMBED_SIZE = 128  # Size of word embeddings
HIDDEN_SIZE = 256  # Number of RNN hidden units

# Instantiate model
model = SimpleRNNModel(VOCAB_SIZE, EMBED_SIZE, HIDDEN_SIZE)

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print(model)
```

```python
import torch.optim as optim

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
import numpy as np

SEQUENCE_LENGTH = 5  # Number of words in input sequence

# Create input-output pairs
input_sequences = []
target_words = []

for i in range(len(tokenized_text) - SEQUENCE_LENGTH):
    input_sequences.append(tokenized_text[i:i + SEQUENCE_LENGTH])  # First SEQUENCE_LENGTH tokens
    target_words.append(tokenized_text[i + SEQUENCE_LENGTH])  # The next word as the target

# Convert to PyTorch tensors
input_sequences = torch.tensor(np.array(input_sequences), dtype=torch.long)
target_words = torch.tensor(np.array(target_words), dtype=torch.long)

print("Input sequences shape:", input_sequences.shape)  # Expected: (num_samples, SEQUENCE_LENGTH)
print("Target words shape:", target_words.shape)  # Expected: (num_samples,)
```

```python
from torch.utils.data import TensorDataset, DataLoader

# Define batch size
BATCH_SIZE = 64

# Create dataset
dataset = TensorDataset(input_sequences, target_words)

# Create DataLoader
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

print("DataLoader created successfully!")
```

```python
# Training Loop

NUM_EPOCHS = 5  # Number of training epochs

for epoch in range(NUM_EPOCHS):
    total_loss = 0

    for batch in dataloader:
        inputs, targets = batch
        inputs, targets = inputs.to(device), targets.to(device)

        optimizer.zero_grad()  # Clear gradients
        outputs = model(inputs)  # Forward pass
        loss = criterion(outputs, targets)  # Compute loss
        loss.backward()  # Backpropagation
        optimizer.step()  # Update weights

        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{NUM_EPOCHS}, Loss: {total_loss/len(dataloader):.4f}")
```

```python
import torch.nn.functional as F

def predict_next_word(model, tokenizer, text, top_k=5):
    model.eval()
    with torch.no_grad():
        # Tokenize input text
        input_ids = tokenizer.encode(text, return_tensors="pt").to(device)

        # Get model predictions
        output = model(input_ids)
        probabilities = F.softmax(output, dim=-1)

        # Get top-k predicted words
        top_k_probs, top_k_indices = torch.topk(probabilities, top_k)

        # Convert token IDs to words
        predicted_words = [tokenizer.decode([idx]) for idx in top_k_indices[0]]

        return predicted_words

# Test the prediction function
input_text = "The future of AI is"
predicted_words = predict_next_word(model, tokenizer, input_text)
print("Next word predictions:", predicted_words)
```
