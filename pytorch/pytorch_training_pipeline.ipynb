```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
```

```python
df
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
y_train
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

```python
y_train
```

```python
X_train_tensor=torch.tensor(X_train,dtype=torch.float32)
X_test_tensor=torch.tensor(X_test,dtype=torch.float32)
y_train_tensor=torch.tensor(y_train,dtype=torch.long)
y_test_tensor=torch.tensor(y_test,dtype=torch.long)
```

```python
X_train_tensor
```

```python
class MySimpleNN():
  def __init__(self,X):
    self.weights=torch.randn(X.shape[1],1,requires_grad=True)
    self.bias=torch.randn(1,requires_grad=True)
  def forward(self,X):
    return torch.sigmoid(torch.matmul(X,self.weights)+self.bias)
  def loss_function(self,y_pred,y_true):
    return torch.nn.functional.binary_cross_entropy(y_pred,y_true.unsqueeze(1).float())


```

```python
learning_rate=0.1
epochs=25
```

```python
model=MySimpleNN(X_train_tensor)
model.weights
```

```python
for epoch in range(epochs):
  y_pred=model.forward(X_train_tensor)
  loss=model.loss_function(y_pred,y_train_tensor)
  loss.backward()
  with torch.no_grad():
    model.weights-=learning_rate*model.weights.grad
    model.bias-=learning_rate*model.bias.grad
    model.weights.grad.zero_()
    model.bias.grad.zero_()
  print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item()}')
```

```python
model.weights
```

```python
with torch.no_grad():
  y_pred=model.forward(X_test_tensor)
  y_pred_class=torch.where(y_pred>0.5,1,0)
  accuracy=(y_pred_class==y_test_tensor).sum().item()/len(y_test_tensor)
  print(f'Accuracy: {accuracy}')
```
